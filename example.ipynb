```python
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key="<OPENROUTER_API_KEY>",
)

completion = client.chat.completions.create(
  extra_headers={
  },
  extra_body={},
  model="qwen/qwen3-32b:free",
  messages=[
    {
      "role": "user",
      "content": prompt_polars
    }
  ],
  temperature=0.0,
)

print(completion.choices[0].message.content)
```

```python
df = pl.read_csv('data/freelancer_earnings_bd.csv')
```

```python
pl.sql('SELECT Client_Region, SUM(Earnings_USD) FROM df GROUP BY Client_Region').collect()
```

```python
pl.sql("SELECT ROUND((COUNT(CASE WHEN Experience_Level = 'Expert' AND Job_Completed < 100 THEN 1 END) * 100.0 / COUNT(CASE WHEN Experience_Level = 'Expert' THEN 1 END)), 2) AS percentage FROM df WHERE Experience_Level = 'Expert'").collect()
```
